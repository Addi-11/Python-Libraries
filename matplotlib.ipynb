```python
from matplotlib import pylab
import numpy as np
```

## Plot Graphs

```python
x = np.linspace(0,10,25)
y = x*x + 2
pylab.plot(x,y,'r*-') # 'r' for red
```

```python
## Drawing Sub-Graph

pylab.subplot(1,2,1)# rows columns indexes
pylab.plot(x,y,'r--')
pylab.subplot(2,2,2)
pylab.plot(y,x,'g*-')

#combining both graphs
pylab.subplot(1,2,1)
pylab.plot(y,x,'b*-') ## to chk odd or even funct with graph
```

## Operation Description

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.9,0.1,1,1]) #Control the left, right , width, height of the canvas from 0 to 1
axis.plot(x,y,'r')
```

```python
# Drawing subplots using pyplot
fig, axes = plt.subplots(nrows = 1, ncols = 2) # Submap is of 1 row,2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
# drawing graph or picture inside another graph
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize = (16,9), dpi = 300) # new graphic object
#figsize to adjust the size across the screen 
#dpi(dots per inches) to ensure qualilty while printing or displaying on big screens

fig.add_subplot()
plt.plot(x,y,'r')
```

```python
# instead of plot axes too can be used
fig, axes = plt.subplots(figsize = (16,9), dpi = 300)
#skipping the figsize arg will give std size
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("Title Of Graph")
axes.plot(x,y,'r')
```

```python
# Different plot on the same graph
ax.legend(["label1","label2","label3"])

fig,axes = plt.subplots(dpi = 100)
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")

axes.set_title("Title Of Graph")

axes.plot(x, x**2)
axes.plot(x, x**3)
axes.plot(x, 200*np.log(x))

axes.legend(["y = x^2", "y = x^3","y = 200*log(x)"], loc = 0)
# loc = i corresponds to different locations on the screen
```

## Graph Customization

```python
fig, ax = plt.subplots(dpi = 100)

ax.plot(x, x+2, color = 'red', lw = 0.5, alpha = 0.5) # lw for line width
ax.plot(x, 3*x, color = 'blue', alpha = 0.3)
ax.plot(x, np.log(x), lw = 1.25)
ax.plot(x, x*2)
```

```python
fig, ax = plt.subplots(dpi = 100)

ax.plot(x, x+1, color = 'blue', lw = 2, linestyle='-')
ax.plot(x, x+2, color = 'blue', lw = 2, linestyle='-.')
ax.plot(x, x+3, color = 'blue', lw = 2, linestyle=':')

line = ax.plot(x,x+4, color = 'black', lw = 1.5, dashes = ([5,10]))

line = ax.plot(x, x+5, color = 'black', lw = 1.5)
line.set_dashes([5,10,4,10])
```

```python
fig, ax = plt.subplots(dpi = 100)

ax.plot(x, x+1, color = 'blue', marker = '+')
ax.plot(x, x+2, color = 'blue', marker = 'o')
ax.plot(x, x+3, color = 'blue', marker = 's')# s for square
ax.plot(x, x+4, color = 'blue', marker = '1')
ax.plot(x, x+5, color = 'blue', marker = 'o', markersize = 4)
ax.plot(x, x-1, color = 'blue', marker = 'v', markerfacecolor = 'red')
ax.plot(x, x, color = 'blue', marker = 's',markersize = 4, markerfacecolor = 'yellow')
```

```python
# set canvas grid and axis

fig, Axes = plt.subplots(1,2, figsize = (10,5))

Axes[0].plot(x, x**2, x, x**3, lw = 2)
Axes[0].grid(True)

Axes[1].grid(True)
Axes[1].plot(x, x**2, x, x**3, lw = 2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
```

## Other 2-D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize = (20,6))

axes[0].set_title("scatter")
axes[0].scatter(x,x + 0.5*np.random.rand(len(x)))

axes[1].set_title("step")
axes[1].step(n,n**2, lw = 2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align = 'center', alpha = 0.5)

axes[3].set_title("fill-inbetween")
axes[3].fill_between(x,x**2,x**3, alpha = 0.5) # like integration
```

```python
## Draw Radar Chart

fig = plt.figure(figsize = (6,6))
ax = fig.add_axes([0.0,0.0,1,1], polar = True)

t = np.linspace(0,2*np.pi, 100)
ax.plot(t,2*t, color = 'blue', lw = 3)
```

```python
# Draw Histogram

n = np.random.rand(100)
fig, axes = plt.subplots(1,2, figsize = (12,4))

axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, cumulative = True, bins = 50)
```

```python
# Draw Contour Image

import matplotlib.cm as cm # color maps

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)

X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1-Z2)*2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title('Contour Map')
```

## 3-D Surface Images

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (14,6))

ax = fig.add_subplot(1,1,1, projection = '3d')
ax.plot_surface(X,Y,Z, rstride = 4, cstride = 4, lw = 0)
```

### Pie Chart to represent popularity of all Languages

```python
#Data to plot

languages = 'Python', 'Java', 'PHP', 'JavaScript','C#','C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4",'#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b']
```

```python
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, explode = explode, labels = languages, colors = colors, autopct = '%1.1f%%',shadow = True, startangle = 90)
```
